```python
plt.figure(figsize=(16, 15))

plt.subplot(4,4,1)
sns.distplot(df['clock_speed'], color='brown')

plt.subplot(4,4,2)
sns.distplot(df['fc'], color='green')

plt.subplot(4,4,3)
sns.distplot(df['m_dep'], color='cyan')

plt.subplot(4,4,4)
sns.distplot(df['n_cores'], color='orange')

plt.subplot(4,4,5)
sns.distplot(df['pc'], color='blue')

plt.subplot(4,4,6)
sns.distplot(df['sc_h'], color='black')

plt.subplot(4,4,7)
sns.distplot(df['sc_w'], color='lime')

plt.subplot(4,4,8)
sns.distplot(df['talk_time'], color='red')
```

```python
plt.figure(figsize=(16, 16))

plt.subplot(7,7,1)
labels = 'Yes', 'No'
colors = ['lightblue', 'white']
plt.pie(df['blue'].value_counts(), labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=0)
plt.axis('equal')
plt.title('Bluetooth')

plt.subplot(7,7,2)
labels = 'Yes', 'No'
colors = ['lightgreen', 'white']
plt.pie(df['dual_sim'].value_counts(), labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=0)
plt.axis('equal')
plt.title('Dual Sim')

plt.subplot(7,7,3)
labels = 'Yes', 'No'
colors = ['pink', 'white']
plt.pie(df['three_g'].value_counts(), labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=0)
plt.axis('equal')
plt.title('3G')

plt.subplot(7,7,4)
labels = 'Yes', 'No'
colors = ['cyan', 'white']
plt.pie(df['four_g'].value_counts(), labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=0)
plt.axis('equal')
plt.title('4G')

plt.subplot(7,7,5)
labels = 'Yes', 'No'
colors = ['brown', 'white']
plt.pie(df['touch_screen'].value_counts(), labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=0)
plt.axis('equal')
plt.title('Touch Screen')

plt.subplot(7,7,6)
labels = 'Yes', 'No'
colors = ['lightgrey', 'white']
plt.pie(df['wifi'].value_counts(), labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=0)
plt.axis('equal')
plt.title('wifi')

plt.subplot(7,7,7)
labels = 1,2,3,4
colors = ['lightgreen', 'lightblue', 'pink', 'lightyellow']
plt.pie(df['price_range'].value_counts(), labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=0)
plt.axis('equal')
plt.title('Price Class')

plt.show()
```
